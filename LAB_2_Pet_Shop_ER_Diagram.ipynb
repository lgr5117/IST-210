#Explanation:

In our Pet Store ER Diagram, we added six entities: Pets, Enclosure, Food, Pet Transactions, Customer, and Food Transactions. The pet entity has a primary key named petID and four attributes: type, cost, DOB (date of birth) and a foreign key called enclosure. The foreign key "enclosure" connects to the primary key, "enclosureID" in the entity "Enclosure". These two entities conected through "one to one enclosure kept zero to ten pets." The number ten was chosen as our limit because such enclosure could be an aquarium holding ten or less fish. The attributes in the entity "Enclosure" are size, pet, and cost. Going back to the entity "Pets", the primary key "petID" connects to foreign key "pets" in the "Pet Transactions" entity. This is connected through "one to one petID is sold to one to ten pets." We chose the maximum number ten again to represent if someone were to purchase ten or less fish for example. The other attributes in the entity "Pet Transactions" include totalPrice, accessories, and another foreign key "customerID", which is connected to entity "Customer." This is connected together in the sense that "one customer purchases and makes one to one thousand pet transactions to many customerIDs per year." The number one thousand is obviously high for adopting pets, but our entity "pet transactions" attributes also take account of a total price and extra accessories bought for such pet(s). Our entity "Customer" with a primary key of customerID contains the attributes "firstName", "lastName", "DOB", "loyaltyID", and "transactions." Our primary key "customerID" connects to the entity "food transactions'" foreign key attribute "customer." This connects  by one to one customerID purchases/makes one to one thousand transactions by many customers per year. The entity "food transactions" has the attributes (foreign key) "customer", "totalPrice", and (foreign key)"foodID." The foreign key "foodID" connects the entity "Food's" primary key "foodID." This is represented as "one to one foodID is sold by one to ten types of food (foodID)." We chose the number ten as a limit because we believed no more than ten different types of food would be purchased in one transaction. Our last entity, "Food" has a primary key labeled "foodID" and three attributes: foodType, cost, and size.
