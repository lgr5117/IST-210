#Question 1: What is the average cost of all the Vehicle Types?

```
%%sql
SELECT AVG(cost) from VehicleType
```

**Why this question is important:**
This question is important because it shows consumers what the average cost of all vehicles are, and furthermore, buyers will know if they have enough money to rent from our business. It will also help our business and advertisement because we will we able to tell people what the price range which will represent the affordableness our service. Also, our business can either increase or decrease the renting price to satisfy renters. If our price is to expensive and our sales are low, we can make the correct adjustment to be more appealing. Yes, the results did match up from our expectations because it added up the cost of each vehicle and then divided it by the total amount of vehicles to give potntial renters the price range of our business.

#Question 2: How many of each type of payment method are there?

The results matched up with our expectations because it outputted how many payment methods are available for our business.

#Question 3: What is the total rental cost of each vehicle type? 

```
%%sql
SELECT Type, SUM(Cost) from VehicleType GROUP BY Type ORDER BY 2 DESC
```

**Why this question is important:** This question is important because it displays the average rental cost for each type of vehicle. This question is more specific from question one because it helps display the average cost for the vehicle that the customer is interested in. If the customer wants to rent a motorcycle, they will be able to see how much, on average, it will cost them. This is helpful for our business because it can show us reasons why some types of vehicles arent selling. For example, if ATVs arent getting renting a lot, we can decrease the renting price so it is more appealing to the customers. As well, this is a way to keep track of the monetary inventory value overall. The results matched up with our expectations because we are able to tell the difference in renting cost for each type of vehicle.
